```python
data['hotel'] = data['hotel'].astype('category')
data['meal'] = data['meal'].astype('category')
data['market_segment'] = data['market_segment'].astype('category')
```

```python
data = data.drop_duplicates()
```

```python
data = data[data['adr']>0]
```

```python
data.head()
```

```python
data['total_guests'] = data['adults'] + data['children'] + data['babies']
```

```python
data['total_stay'] = data['stays_in_weekend_nights'] + data['stays_in_week_nights']
```

```python
data['previous_cancellations'].value_counts()
```

```python
data['repeat_ratio'] = data['previous_bookings_not_canceled']/(data['previous_cancellations']+1)
data['repeat_ratio'].value_counts()
```

```python
data['lead_time'] = pd.cut(data['lead_time'],
                         bins = [0,30,90,180,365,data['lead_time'].max()],
                         labels = ['short','mid','long','very_long','extreme'],
                         include_lowest = True)
```

```python
data['lead_time'].value_counts()
```

```python
data['lead_deposit_interaction'] = data['lead_time'].astype(str)+'_'+data['deposit_type']
```

```python
data['adr_bin'] = pd.cut(data['adr'],
                         bins = [0,50,100,200,data['adr'].max()],
                         labels = ['low','medium','high','very_high'],
                         include_lowest=True)
```

```python
data['adr_bin'].value_counts()
```

```python
seasonal_avg = data.groupby('arrival_date_month')['adr'].mean()
seasonal_avg
```

```python
data['seasonal_adr'] = data.apply(
    lambda row: row['adr'] - seasonal_avg[row['arrival_date_month']],
    axis=1
)
```

```python
data.rename(columns={'arrival_date_month': 'arrival_month_num'}, inplace=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=['is_canceled'])
y = data['is_canceled']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.22,random_state=42,stratify=y)
```

```python
numeric_features = [
    'lead_time',
    'arrival_month_num',
    'days_in_waiting_list',
    'adr',
    'total_of_special_requests',
    'total_guests',
    'total_stay',
    'repeat_ratio',
    'seasonal_adr',
    'stability'
]

ordinal_features = ['lead_time_bin']

nominal_features = [
    'hotel',
    'deposit_type',
    'lead_deposit_interaction'
]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier

numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

ordinal_transformer = Pipeline(steps=[
    ('ordinal', OrdinalEncoder(categories=[['short', 'mid', 'long', 'very_long', 'extreme']]))
])

nominal_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('ord', ordinal_transformer, ordinal_features),
        ('nom', nominal_transformer, nominal_features)
    ]
)
```

```python
xgb_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(
        objective='binary:logistic',
        eval_metric='logloss',
        n_estimators=300,
        learning_rate=0.1,
        max_depth=6,
        subsample=0.9,
        colsample_bytree=0.9,
        random_state=42,
        n_jobs=-1,
        reg_lambda=1.0
    ))
])
```

```python
xgb_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# predictions & probs
y_pred = xgb_model.predict(X_test)
y_proba = xgb_model.predict_proba(X_test)[:, 1]  # probability for positive class

# metrics
print("F1:", f1_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("\nClassification report:\n", classification_report(y_test, y_pred))

# confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title('Confusion Matrix')
plt.show()

# ROC AUC
auc = roc_auc_score(y_test, y_proba)
fpr, tpr, _ = roc_curve(y_test, y_proba)
print("ROC AUC:", auc)
plt.figure()
plt.plot(fpr, tpr, label=f"AUC={auc:.3f}")
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel('FPR'); plt.ylabel('TPR'); plt.title('ROC Curve'); plt.legend()
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBClassifier

param_dist = {
    'classifier__n_estimators': [200, 300, 400, 500],
    'classifier__learning_rate': [0.01, 0.05, 0.1, 0.2],
    'classifier__max_depth': [4, 5, 6, 8],
    'classifier__min_child_weight': [1, 3, 5],
    'classifier__subsample': [0.7, 0.8, 0.9],
    'classifier__colsample_bytree': [0.6, 0.8, 1.0],
    'classifier__gamma': [0, 0.1, 0.2, 0.4],
    'classifier__reg_lambda': [1, 1.5, 2],
    'classifier__scale_pos_weight': [1, 2, 3]  # helps recall massively
}


xgb_random = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=20,                # FAST calibration (5–8 mins)
    scoring='f1',
    cv=3,
    verbose=2,
    random_state=42,
    n_jobs=1
)


xgb_random.fit(X_train, y_train)

print("Best Parameters:")
print(xgb_random.best_params_)

print("\nBest F1 (CV):", xgb_random.best_score_)
```

```python
best_xgb = xgb_random.best_estimator_

y_pred = best_xgb.predict(X_test)
y_proba = best_xgb.predict_proba(X_test)[:, 1]

print("Tuned F1:", f1_score(y_test, y_pred))
print("Tuned Precision:", precision_score(y_test, y_pred))
print("Tuned Recall:", recall_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
```

```python
y_pred = xgb_model.predict(X_test)
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import precision_recall_curve, f1_score
import numpy as np

y_scores = best_xgb.predict_proba(X_test)[:, 1]

prec, rec, thresholds = precision_recall_curve(y_test, y_scores)

f1_scores = 2 * (prec * rec) / (prec + rec + 1e-6)
best_idx = np.argmax(f1_scores)

best_threshold = thresholds[best_idx]
best_f1 = f1_scores[best_idx]

print("Best Threshold:", best_threshold)
print("Best F1 Score:", best_f1)
```

Focal Loss implementation for xgboost

```python
import numpy as np

def focal_loss(alpha=0.25,gamma=2):
    def fl_obj(y_true,y_pred):
        p=1.0/(1.0+np.exp(-y_pred))
        grad = alpha * (y_true - p) * ((1 - p) ** gamma)

        # hessian
        hess = alpha * ((1 - p) ** gamma) * p * (1 - p) * \
               (1 + gamma * (p - y_true))

        return -grad, hess  # NEG grad so XGBoost minimizes the loss

    return fl_obj
```

```python
def focal_loss_eval(alpha=0.25, gamma=2.0):
    def fl_eval(y_pred, dtrain):
        y_true = dtrain.get_label()
        p = 1.0 / (1.0 + np.exp(-y_pred))

        fl = -alpha * (y_true * (1 - p)**gamma * np.log(p + 1e-7) +
                       (1 - y_true) * (p**gamma) * np.log(1 - p + 1e-7))

        return 'focal_loss', np.mean(fl)
    return fl_eval
```

```python
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline

xgb_focal = XGBClassifier(
    objective=focal_loss(alpha=0.25, gamma=2.0),  # custom loss
    eval_metric=focal_loss_eval(alpha=0.25, gamma=2.0),
    n_estimators=400,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.9,
    tree_method="hist",    # faster on mac
    random_state=42,
    n_jobs=1
)
```

```python
model_focal = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', xgb_focal)
])
```

```python
model_focal.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

y_pred = model_focal.predict(X_test)
y_proba = model_focal.predict_proba(X_test)[:, 1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1:", f1_score(y_test, y_pred))
print("AUC:", roc_auc_score(y_test, y_proba))
```

```python
categorical_cols = [
    'hotel',
    'deposit_type',
    'lead_deposit_interaction',
    'lead_time_bin',
    'guest_category'
]
```

```python
from catboost import CatBoostClassifier

cat_model = CatBoostClassifier(
    iterations=1000,
    learning_rate=0.05,
    depth=8,
    loss_function='Logloss',
    eval_metric='F1',
    random_seed=42,
    l2_leaf_reg=5,
    border_count=128,
    verbose=200
)
```

```python
cat_model.fit(
    X_train,
    y_train,
    cat_features=categorical_cols,
    eval_set=(X_test, y_test),
    use_best_model=True
)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

y_pred = cat_model.predict(X_test)
y_proba = cat_model.predict_proba(X_test)[:, 1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1:", f1_score(y_test, y_pred))
print("AUC:", roc_auc_score(y_test, y_proba))
# Accuracy: 0.7717397078353254
# Precision: 0.66078697421981
# Recall: 0.3714013346043851
# F1: 0.4755278896619065
# AUC: 0.7828088466328125
```

CatBoost with class weights

```python
from collections import Counter

counter = Counter(y_train)
neg = counter[0]
pos = counter[1]

weight_0 = 1
weight_1 = neg / pos   # minority gets higher weight

print("Class ratio:", neg, pos)
print("Weight for class 1:", weight_1)
```

```python
from catboost import CatBoostClassifier

cat_model_weighted = CatBoostClassifier(
    iterations=6000,        # long training (good with class weights)
    learning_rate=0.03,
    depth=8,
    loss_function='Logloss',
    eval_metric='F1',
    random_seed=42,
    l2_leaf_reg=5,
    border_count=128,
    class_weights=[weight_0, weight_1],
    verbose=400
)
```

```python
cat_model_weighted.fit(
    X_train,
    y_train,
    cat_features=categorical_cols,
    eval_set=(X_test, y_test),
    use_best_model=True
)
```
